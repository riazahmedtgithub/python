```python

import scrapy
from scrapy.crawler import CrawlerProcess

#Define the class
class DC_Description_Spider(scrapy.Spider):
  name='myspider'
#Define the initial mandatory start_requests method
  def start_requests(self):
    yield scrapy.Request(url = "https://cloud.google.com/iot/pricing", callback = self.first_parse)
#Call the first website through the first parse
  def first_parse(self,response):
    res=response.xpath("//td[1]/text() | //td[2]/text()").extract()
#Convert the list to dictionary format
    costcompare1={res[i]:res[i+1] for i in range(0,len(res),2)}
    print("Output of iot pricing in dictionary format", costcompare1)
#Get the second value by using indexing
    self.cost1=list(costcompare1.values())[1]
#Get the value of cost1 from the first website
    print("Output of one of the iot pricing", self.cost1)
    yield response.follow(url = "https://cloud.google.com/storage/pricing", callback = self.second_parse)
#Call the second website through the second parse
  def second_parse(self,response):
    res=response.xpath('//tr[1]/td[1]/text() | //tr[1]/td[2]/text()').extract()
    #Getting the cost which is 3rd value from the list
    cost2=res[2]
#Get the value of cost2 from the second website
    print("Output of one of the storage pricing", cost2)
#Call the compareanddisplay method to compare the costs
    self.compareanddisplay(self.cost1,cost2)
#compareanddisplay method
  def compareanddisplay(self,cost1,cost2):
    if self.cost1>cost2:
      print("IOT pricing is more")
    else:
      print("Storage pricing is more")
   
process = CrawlerProcess()
process.crawl(DC_Description_Spider)
process.start()

```
