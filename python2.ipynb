```python

import random
i=0
lis=[]
lisnum=[]
lisspl=[]
while i < 3:
    a=random.randint(ord('a'),ord('z'))
    b=random.randint(1,9)
    lisnum.append(str(b))
    char=chr(a)
    lis.append(char)
    lisspl.append(random.choices("()*&^;:\/?"))
    i=i+1
lisspl1=[str(j) for i in lisspl for j in i]
lis=lis+lisnum+lisspl1
print(''.join(random.sample(lis,len(lis))))
```
